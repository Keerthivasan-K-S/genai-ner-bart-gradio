```python
from transformers import pipeline
import gradio as gr
```

```python
ner_pipeline = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)
```

```python
def recognize_entities(text):

    results = ner_pipeline(text)

    if len(results) == 0:
        return "No entities found."

    output = ""

    for entity in results:
        output += (
            f"Entity: {entity['word']}\n"
            f"Type: {entity['entity_group']}\n"
            f"Confidence: {round(entity['score'],4)}\n\n"
        )

    return output
```

```python
interface = gr.Interface(
    fn=recognize_entities,
    inputs=gr.Textbox(
        lines=5,
        placeholder="Enter text here..."
    ),
    outputs="text",
    title="Named Entity Recognition using Fine-Tuned Model",
    description="Extracts people, organizations, and locations from text."
)
```

```python
interface.launch()
```
